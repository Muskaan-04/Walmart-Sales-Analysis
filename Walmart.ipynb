```python
df=pd.read_csv('Walmart.csv')
```

```python
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['promotion_type'].fillna('No Promotion', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['transaction_date']=pd.to_datetime(df['transaction_date'], format='mixed')
```

```python
df['transaction_date'].dt.date
```

```python
df['transaction_date'].dt.time
```

```python
df['total_revenue']=df['quantity_sold']*df['unit_price']
df['total_revenue']
```

```python
daily_sales = df.groupby(df['transaction_date'].dt.date)['total_revenue'].sum()
print(daily_sales)
```

```python
weekly_sales=df.groupby(df['transaction_date'].dt.isocalendar().week)['total_revenue'].sum()
print(weekly_sales)
```

```python
monthly_sales=df.groupby(df['transaction_date'].dt.month)['total_revenue'].sum()
print(monthly_sales)
```

```python
df['forecast_error']=abs(df['actual_demand']-df['forecasted_demand'])
```

```python
df['forecast_error']
```

```python
print("total forecasted_error :",df['forecast_error'].sum())
```

```python
print("avg forcast_error :",df['forecast_error'].mean())
```

```python
print("max forecast_error :",df['forecast_error'].max())
```

```python
print("min forecasted_error :",df['forecast_error'].min())
```

```python
promotion_sales=df[df['promotion_applied']==True]['total_revenue'].sum()
```

```python
print(promotion_sales)
```

```python
nonpromotion_sales=df[df['promotion_applied']==False]['total_revenue'].sum()
```

```python
print(nonpromotion_sales)
```

```python
category_sales=df.groupby('category')['total_revenue'].sum()
print(category_sales)
```

```python
quantity_sales=df.groupby('quantity_sold')['total_revenue'].sum()
print(quantity_sales)
```

```python
product_sales=df.groupby('product_name')['total_revenue'].sum()
print(product_sales)
```

```python
payment_type=df.groupby('payment_method')['total_revenue'].sum()
print(payment_type)
```

```python
store_sales =df.groupby('store_location')['total_revenue'].sum()
print(store_sales)
```

```python
customer_info=df.groupby(['customer_age','customer_income','customer_gender'])['total_revenue'].sum()
print(customer_info)
```

```python
loyalty_level=df.groupby('customer_loyalty_level')['total_revenue'].sum()
print(loyalty_level)
```

```python
weather_sales=df.groupby('weather_conditions')['total_revenue'].sum()
print(weather_sales)
```

```python
Wcondtion_sales=df.groupby(['weather_conditions','category'])['total_revenue'].sum()
print(Wcondtion_sales)
```

```python
daily_sales.to_csv("daily_sales.csv")
weekly_sales.to_csv("weekly_sales.csv")
monthly_sales.to_csv("monthly_sales.csv")
category_sales.to_csv("category_sales.csv")
product_sales.to_csv("product_sales.csv")
store_sales.to_csv("store_sales.csv")
payment_type.to_csv("payment_type.csv")
weather_sales.to_csv("weather_sales.csv")
loyalty_level.to_csv("loyalty_sales.csv")
```
